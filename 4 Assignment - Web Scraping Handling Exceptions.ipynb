```python
start = 0
end = 3

for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
    for i in url:
        Job_opening_url.append(i.get_attribute('href'))
    
    next_butt=driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/div/div/section[2]/div[3]/div/a[2]/span')
```

```python
current_page = 1
while current_page <= 3:
    products = driver.find_elements(By.XPATH, "//div[@data-component-type='s-search-result']")

    for product in products:
        try:
            brand_name = product.find_element(By.XPATH, ".//span[@class='a-size-base-plus a-color-base a-text-normal']").text
        except:
            brand_name = '-'
        
        try:
            product_name = product.find_element(By.XPATH, ".//h2/a/span[@class='a-size-medium a-color-base a-text-normal']").text
        except:
            product_name = '-'
        
        try:
            price = product.find_element(By.XPATH, ".//span[@class='a-price-whole']").text
        except:
            price = '-'
        
        try:
            return_exchange_info = product.find_element(By.XPATH, ".//span[contains(text(), 'Return')]").text
        except:
            return_exchange_info = '-'
        
        try:
            expected_delivery_info = product.find_element(By.XPATH, ".//span[contains(text(), 'FREE Delivery')]").text
        except:
            expected_delivery_info = '-'
        
        try:
            availability = product.find_element(By.XPATH, ".//span[@class='a-size-medium a-color-success']").text
        except:
            availability = '-'
        
        try:
            product_url = product.find_element(By.XPATH, ".//h2/a").get_attribute("href")
        except:
            product_url = '-'

        try:
            next_page = driver.find_element(By.XPATH, "//li[@class='a-last']/a")
            next_page.click()
            time.sleep(5) 
            current_page += 1
        except:
            break

driver.quit()
```

```python
        
brand_names.append(brand_name)
product_names.append(product_name)
prices.append(price)
return_exchange_info_list.append(return_exchange_info)
expected_delivery_info_list.append(expected_delivery_info)
availability_list.append(availability)
product_urls.append(product_url)
    
df = pd.DataFrame({'Brand Name': brand_names,'Name of the Product': product_names,'Price': prices,
    'Return/Exchange': return_exchange_info_list,'Expected Delivery': expected_delivery_info_list,
    'Availability': availability_list,'Product URL': product_urls})
```

```python
try:
    url= driver.current_url
    print("URL Extracted: ", url)
    lat_lng = re.findall(r'@(.*)data',url)
    if len(lat_lng):
        lat_lng_list = lat_lng[0].split(",")
        if len(lat_lng_list)>=2:
            lat = lat_lng_list[0]
            lng = lat_lng_list[1]
        print("Latitude = ",lat, "Longitude = ",lng)

except Exception as e:
        print("Error: ", str(e))
        
driver.close()
```
